```python
import pandas as pd
import numpy as np
df=pd.read_csv("Sales_Satisfication_with_NaNs.csv")
df
```

```python
df.isnull().sum()
df[df['Group'].isnull()]
df.isna().sum()
```

```python
df['Group'].value_counts()
```

```python
df['Customer_Segment'].value_counts()
```

```python
df['Group']=df['Group'].replace('nan',np.nan).fillna('unknown').str.strip().str.title()
df['Group']
```

```python
# handling empty custumer segment

df['Customer_Segment']=df['Customer_Segment'].replace('nan',np.nan).fillna('not given').str.strip().str.title()
df['Customer_Segment']
```

```python
# handling empty purchase made

df['Purchase_Made']=df['Purchase_Made'].replace('nan',np.nan).fillna('unknown').str.strip().str.title()
df['Purchase_Made']
```

```python
# filling empty values

average_before_sales=df['Sales_Before'].mean()
average_after_sales=df['Sales_After'].mean()
average_before_satisfaction=df['Customer_Satisfaction_Before'].mean()
average_after_satisfaction=df['Customer_Satisfaction_After'].mean()
```

```python
df['Sales_Before']=df['Sales_Before'].fillna(average_before_sales)
df['Sales_Before']
```

```python
df['Sales_After']=df['Sales_After'].fillna(average_after_sales)
df['Sales_After']
```

```python
df['Customer_Satisfaction_After']=df['Customer_Satisfaction_After'].fillna(average_after_satisfaction)
df['Customer_Satisfaction_After']
```

```python
df['Customer_Satisfaction_Before']=df['Customer_Satisfaction_Before'].fillna(average_before_satisfaction)
df['Customer_Satisfaction_Before']
```

```python
# handling negative values

df['Sales_Growth']=df['Sales_Growth'].abs()
df['Satisfaction_Growth']=df['Satisfaction_Growth'].abs()
df['ROI_index']=df['ROI_index'].abs()
```

```python
df=df.fillna('')
```

```python
df.to_csv('cleaned_sales_satisfaction.csv',index=False)
```
